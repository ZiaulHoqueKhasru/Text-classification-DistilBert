```python
from sagemaker.huggingface import HuggingFaceModel
from sagemaker import get_execution_role

role = get_execution_role()
model_s3_path = 's3://sagemaker-text-classifier-uci-news-aggregation/output/huggingface-pytorch-training-2026-01-13-17-38-35-955/output/model.tar.gz'
#model_s3_path = 's3://hugging-face-multiclass-textclassification-bucket/output/huggingface-pytorch-training-2024-07-08-08-10-28-142/output/model.tar.gz'
huggingface_model = HuggingFaceModel(

    model_data = model_s3_path,
    role = role,
    transformers_version = '4.48.0',#in deployment found this stable after several try
    pytorch_version= '2.3.0',#this is must in deployment end of 2025
    py_version = 'py311',
    entry_point = 'inference.py',
    name = "load-testing-model" #the model will be created by this name in aws models section  
)
```

# after finising i must delete the end point
sagemaker -> Deployment and Inference -> endpoint -> delele

```python
predictor = huggingface_model.deploy(
    initial_instance_count= 1,
    instance_type= 'ml.m5.xlarge',#ml.t3.xlarge was not supporting
    endpoint_name= 'multiclass-text-classification-endpoint-ziaV3'#after delteing this i need to set another new name when re run it,because it requres unique endpoint name
)
```

```python
data = {"inputs": "Covid19 is dengerious"}
prediction = predictor.predict(data)
prediction
```
